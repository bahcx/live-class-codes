    b)	(1 point) With a for loop, draw and study the values in each of the remaining W columns. (hint for each w column draw histogram and box plot)

```python
# Your code
# Select only the weekly ranking columns (those starting with 'w')
week_cols = [col for col in bb_df.columns if col.startswith('w')]

# Loop through each week column and plot
for col in week_cols:
    plt.figure(figsize=(12, 4))
    
    # Histogram
    plt.subplot(1, 2, 1)
    bb_df[col].dropna().hist(bins=20, edgecolor='black')
    plt.title(f'Histogram of {col}')
    plt.xlabel('Chart Position')
    plt.ylabel('Frequency')
    
    # Box plot
    plt.subplot(1, 2, 2)
    bb_df.boxplot(column=col)
    plt.title(f'Boxplot of {col}')
    
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import pandas as pd
gpaDf =  pd.read_excel("https://www.dropbox.com/scl/fi/24jfn2u4rsn6e45bvzkrs/GPA.xlsx?rlkey=9ecqtc90bmntcsucdy8gzyt24&dl=1")
gpaDf
```

    a) (2points) Find regression model equation for the model: GPA = b0 + B1 GRE

```python
# https://www.statsmodels.org/stable/index.html
import statsmodels.api as sm
# Your answer
# Define X (independent variable) and y (dependent variable)
X = gpaDf["GRE"]
y = gpaDf["GPA"]

# Add constant for intercept
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the summary
print(model.summary())
```
